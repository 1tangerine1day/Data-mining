# Build PrefixSpan function

```python
def buildPrefixSpan(data, minSupport):
    
    freqSequences = prefixSpan(SequencePattern([], None, PLACE_HOLDER), data, minSupport)
    freqSeqs = FreqSequences(freqSequences)
    
    return freqSeqs
```

```python
def prefixSpan(pattern, S, threshold):
        patterns = []

        if checkPatternLengths(pattern):
            f_list = frequent_items(S, pattern, threshold)

            for i in f_list:
                p = SequencePattern(pattern.sequence, pattern.freq, PLACE_HOLDER)
                p.append(i)
                if checkPatternLengths(pattern):
                    patterns.append(p)

                p_S = build_projected_database(S, p)
                p_patterns = prefixSpan(p, p_S, threshold)
                patterns.extend(p_patterns)

        return patterns
```

```python
def checkPatternLengths(pattern):
    for s in pattern.sequence:
        if len(s)>maxPatternLength:
            return False
    return True
```

```python
class SequencePattern:
        def __init__(self, sequence, support, place_holder):
            self.place_holder = place_holder
            self.sequence = []
            for s in sequence:
                self.sequence.append(list(s))
            self.freq = support
        def append(self, p):
            if p.sequence[0][0] == self.place_holder:
                first_e = p.sequence[0]
                first_e.remove(self.place_holder)
                self.sequence[-1].extend(first_e)
                self.sequence.extend(p.sequence[1:])
            else:
                self.sequence.extend(p.sequence)
                if self.freq is None:
                    self.freq = p.freq
            self.freq = min(self.freq, p.freq)
```

```python
class FreqSequences:
    def __init__(self, fs):
        self.fs = fs
    def collect(self):
        return self.fs
```

```python
def build_projected_database(S, pattern):
        
        p_S = []
        last_e = pattern.sequence[-1]
        last_item = last_e[-1]
        for s in S:
            p_s = []
            for element in s:
                is_prefix = False
                if PLACE_HOLDER in element:
                    if last_item in element and len(pattern.sequence[-1]) > 1:
                        is_prefix = True
                else:
                    is_prefix = True
                    for item in last_e:
                        if item not in element:
                            is_prefix = False
                            break

                if is_prefix:
                    e_index = s.index(element)
                    i_index = element.index(last_item)
                    if i_index == len(element) - 1:
                        p_s = s[e_index + 1:]
                    else:
                        p_s = s[e_index:]
                        index = element.index(last_item)
                        e = element[i_index:]
                        e[0] = PLACE_HOLDER
                        p_s[0] = e
                    break
            if len(p_s) != 0:
                p_S.append(p_s)

        return p_S
    
```

```python
def frequent_items(S, pattern, threshold):
        items = {}
        _items = {}
        f_list = []
        if S is None or len(S) == 0:
            return []

        if len(pattern.sequence) != 0:
            last_e = pattern.sequence[-1]
        else:
            last_e = []
        for s in S:

            is_prefix = True
            for item in last_e:
                if item not in s[0]:
                    is_prefix = False
                    break
            if is_prefix and len(last_e) > 0:
                index = s[0].index(last_e[-1])
                if index < len(s[0]) - 1:
                    for item in s[0][index + 1:]:
                        if item in _items:
                            _items[item] += 1
                        else:
                            _items[item] = 1

            if PLACE_HOLDER in s[0]:
                for item in s[0][1:]:
                    if item in _items:
                        _items[item] += 1
                    else:
                        _items[item] = 1
                s = s[1:]

            counted = []
            for element in s:
                for item in element:
                    if item not in counted:
                        counted.append(item)
                        if item in items:
                            items[item] += 1
                        else:
                            items[item] = 1

        f_list.extend([SequencePattern([[PLACE_HOLDER, k]], v, PLACE_HOLDER)for k, v in _items.items() if v >= threshold])
        f_list.extend([SequencePattern([[k]], v, PLACE_HOLDER)for k, v in items.items() if v >= threshold])

        f_list = [i for i in f_list if checkPatternLengths(i)]

        sorted_list = sorted(f_list, key=lambda p: p.freq)
        return sorted_list
    
```

# Main function

```python
data = transactionData
PLACE_HOLDER = '_'

model = buildPrefixSpan(data, minSupport)
result = model.collect()
    
```

```python
output = []
for fs in result:
    if len(fs.sequence) > 1:
        tempString = ""
        for i in fs.sequence:
            tempString += "{"
            for j in i:
                if len(j) is 1 or j is i[-1]: 
                    tempString += str(j)
                else:
                    tempString += str(j) + ","
            tempString += "}"
                
        tempString += " SUP:" + str(fs.freq)
        print(tempString)
        output += [tempString]
```

# Output to txt

```python
with open("output.txt", "w") as f:
    for i in output:
        if i is not output[-1]:
            f.write(i+" \n")
        else:
            f.write(i)
```
